### Bank Marketing (UCI Machine Learning Repository)

Estes dados estão relacionados a campanhas de marketing, baseadas em ligações telefônicas, de uma instituição bancária portuguesa. O objetivo original na criação deste conjunto de dados era entender se um cliente iria obter um depósito a prazo (variável y) no banco.

As campanhas de marketing foram realizadas por meio de chamadas telefônicas. Frequentemente, foram necessários mais de um contato com o mesmo cliente para avaliar se o produto (depósito a prazo bancário) vai ser obtido ou não.

```python
# Importação das bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Importação dos dados
df = pd.read_csv(r'data\UCI Machine Learning Repository - Bank.csv', sep=';')
```

```python
# Colunas do dataframe
df.columns
```

##### **1. Distribuição de Idade**

```python
# Total de registros na coluna idade
total_idades = df['age'].count()
print(f"Total de registros de idade: {total_idades}")

# 2. Verificando as idades mais frequentes (Top 5)
print("\nIdades mais comuns no banco:")
print(df['age'].value_counts().head(5))

# Estatísticas rápidas (Média, Mínimo, Máximo)
print("\nResumo estatístico da idade:")
print(df['age'].describe())
```

```python
# Plotar histograma com dados da coluna 'age'
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], kde=True, color='skyblue')
plt.title('Distribuição da Idade dos Clientes')
plt.xlabel('Idade')
plt.ylabel('Quantidade de Clientes')
plt.show()
```
